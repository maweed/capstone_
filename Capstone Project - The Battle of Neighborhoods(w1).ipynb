```python
import numpy as np  
import pandas as pd 
from bs4 import BeautifulSoup

import requests

url='https://en.wikipedia.org/wiki/List_of_shopping_centres_in_Norway'
df = pd.read_html(url, header=0, flavor = 'bs4')
df1=df[0]
df1
```

```python
print('the number of the most famous city center in Norway are {}'.format(len(df1)))
```

We can group location 

```python
df1 = df1.groupby(['Location'])['Name'].apply(', '.join).reset_index()

df1
```
